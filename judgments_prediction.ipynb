# Content

In this file, pairwise preprocessed reviews (x and y) are applied to compute the conditional probability and the marginal probability, hence the Generative Estimator for Mutual Information (GEM). The same procedure is applied for the GEM-S (abstract based GEM) metric. For this both the original and fine-tuned versions of Llama-3.1-8B-Instruct are used.
