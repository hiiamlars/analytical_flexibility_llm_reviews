**Dependencies**

```python
import os
from google.colab import userdata
from openai import OpenAI
```

**API-Key**

```python
try:
    openai_api_key = userdata.get('master_thesis_test')
    if not openai_api_key:
        raise ValueError("API Key not found in Colab Secrets.")
except Exception as e:
    print(f"Error loading key: {e}")

client = OpenAI(api_key=openai_api_key)
```

**Setup**

```python
system_message = "You are an expert tasked with evaluating the quality of a review for a Machine Learning paper. Your goal is ..."

user_prompt = "[Paper]"

print("\nSending request to ChatGPT...")
```

```python
try:
    response = client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=[
            {"role": "system", "content": system_message},
            {"role": "user", "content": user_prompt}
        ]
    )

    chat_response = response.choices[0].message.content
    print("\n--- ChatGPT Response ---\n")
    print(chat_response)
    print("\n--------------------------")

except Exception as e:
    print(f"\nAPI Call Failed. This is likely due to an issue with the API key or rate limits.")
    print(f"Error details: {e}")
```
